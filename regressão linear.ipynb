```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
table_1 = pd.read_csv(r"D:\Arquivos\Downloads\cost_revenue_clean.csv")
```

```python
table_1.describe()
```

```python
sns.pairplot(table_1)
```

```python
table_1.describe()
```

```python
X = DataFrame(table_1, columns=["production_budget_usd"])
y = DataFrame(table_1, columns=["worldwide_gross_usd"])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title("Custo dos Filmes x Receita Global")
plt.xlabel("Custo de Produção USD$")
plt.ylabel("Receita USD$")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

Coeficiente de declividade (theta 1)

Média de ganho por filme

```python
regressao = LinearRegression()
regressao.fit(X, y)
regressao.coef_ #theta_1
```

Previsão do orçamento 0

```python
regressao.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regressao.predict(X), color="red", linewidth=3)
plt.title("Custo dos Filmes x Receita Global")
plt.xlabel("Custo de Produção USD$")
plt.ylabel("Receita USD$")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

Variação prevista pelo modelo (explica a porcentagem de sucesso)

```python
regressao.score(X, y)
```

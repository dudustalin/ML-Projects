## Gradiente descendente como função em Python

```python
def grad_desc(derivativa, escolha, multiplicador=0.02, precisao=0.001, max_iter = 300):
    new_x = escolha
    x_ls = [new_x]
    slope_ls = [derivativa(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivativa(previous_x)
        new_x = previous_x - multiplicador * gradient

        x_ls.append(new_x)
        slope_ls.append(g(new_x))

        step_size = abs(new_x -previous_x)
        if step_size < precisao:
            break
 
        #print(step_size)


    return new_x, x_ls, slope_ls

```

# Gradiente descendente em lote com SymPy

```python
# Parametros

multiplicador = 0.1 #taxa de aprendizado
max_iter = 500
params = np.array([1.8, 1.0]) #palpite inicial

for n in range(max_iter):
    gradient_x = diff(f(a, b), a).evalf(subs={a:params[0], b:params[1]})
    gradient_y = diff(f(a, b), b).evalf(subs={a:params[0], b:params[1]})
    gradients = np.array([gradient_x, gradient_y])
    params = params - multiplicador * gradients

# Resultados
print("Valores na matriz de gradientes", gradients)
print("Mínimo ocorre no valor x ", params[0])
print("Mínimo ocorre no valor y ", params[1])
print("Custo ", f(params[0], params[1]))
```

## Derivadas parciais relativas a $\theta_0$ e $\theta_1$

## $$ \frac{\partial MSE}{\partial \theta_0} = 
\frac{2}{n} \sum_{i=0}^n \big( y^{i} -\theta_0 - \theta_1 x^{i} \big)$$ 

## $$ \frac{\partial MSE}{\partial \theta_1} = 
\frac{2}{n} \sum_{i=0}^n \big( y^{i} -\theta_0 - \theta_1 x^{i} \big) \big( x^i \big)$$ 

## MSE e gradiente descendente

```python
## valores de X, valores de y e parâmetros de theta
def grad (x, y, thetas):
    n=len(x)
    t0_slope = -(2/n) * sum(y - thetas[0] - thetas[1] * x)
    #print(t0_slope, "é")
    t1_slope = -(2/n) * sum((y - thetas[0] - thetas[1] * x) * x)
    #print(t1_slope, "i")
    #return np.array([t0_slope[0], t1_slope[0]])
    #return np.append(arr=t0_slope, values=t1_slope)
    return np.concatenate((t0_slope, t1_slope), axis=0)
```
